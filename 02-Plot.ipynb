In questo capitolo introduciamo alcuni metodi per fare dei semplici grafici in python

```python
# importare packages
import numpy as np #operazioni numeriche
import matplotlib.pyplot as plt #grafici
import os
```

### Plottare una funzione

```python
# generare i dati (x = cos(t)  a<t<b, dt=0.1)
a=0
b=2*np.pi # np.pi = pi greco
dt=0.1
t = np.arange(a,b,dt)
x = np.cos(t)
```

```python
t, x
```

```python
# salvare i dati in un file di testo
# Percorso del file
out_file = '/home/fava/jnotebook/notes/data/cosine.txt'
directory = os.path.dirname(out_file)

# Crea la cartella se non esiste
os.makedirs(directory, exist_ok=True)

# Salva i dati
np.savetxt(out_file, np.column_stack([t, x]))
```

```python
# leggere i dati in una nuova matrice
# Prendendo ovviamente tutte le righe della prima e della seconda colonna
inData = np.loadtxt(out_file)
t2 = inData[:,0] # Nota Python comincia da "0"!
x2 = inData[:,1] 
```

```python
t2, x2
```

```python
#fare un grafico dei dati
plt.plot(t2, x2)
```

```python
# alternativa
plt.plot(inData[:,0],inData[:,1])
```

```python
#Attenzione se plottiamo direttamente la matrice otteniamo questo:
plt.plot(inData)
# Python prende le due colonne separatamente e le plotta!
```

```python
#fare il grafico direttamente
plt.plot(t, np.cos(t))
```

```python
#salvare il plot in un file
plt.plot(t, np.cos(t))
filename='figures/cosine.png'
plt.savefig(filename)
plt.show()
```

**Due curve**

```python
plt.rcParams.update({'font.size': 18})
plt.xlabel("Tempo")
plt.ylabel("Posizione")
plt.title("Oscillatore armonico")
plt.plot(t, np.cos(t), label="x1")
plt.plot(t, np.cos(2*t), label="x2")
plt.legend()
```

**Spostare la legenda**

```python
plt.rcParams.update({'font.size': 18})
plt.xlabel("Tempo")
plt.ylabel("Posizione")
plt.title("Oscillatore armonico")
plt.plot(t, np.cos(t), label="x1")
plt.plot(t, np.cos(2*t), label="x2")
plt.legend(loc="upper left")
# oppure "upper right", "center", etc
```

### Esercizio

Considerare le due funzioni: 
f(x) = sin(x)*exp(-x) e
g(x) = cos(x)*exp(-x)

- Generare i dati con le due funzioni in [0,5] e salvarli in due files 
- Fare un grafico che mostri le due funzioni sempre in [0,5]



```python
a=0
b=5
dx=0.1
x = np.arange(a,b,dx)

f = np.sin(x)*np.exp(-x)
g = np.cos(x)*np.exp(-x)

# salvare i dati in un file di testo
# Percorso del file
out_file_f = '/home/fava/jnotebook/notes/data/f.txt'
out_file_g = '/home/fava/jnotebook/notes/data/g.txt'

directory_f = os.path.dirname(out_file_f)
directory_g = os.path.dirname(out_file_g)

# Crea la cartella se non esiste
os.makedirs(directory_f, exist_ok=True)
# Crea la cartella se non esiste
os.makedirs(directory_g, exist_ok=True)

# Salva i dati
np.savetxt(out_file_f, np.column_stack([x, f]))
# Salva i dati
np.savetxt(out_file_g, np.column_stack([x, g]))

# leggere i dati in una nuova matrice
# Prendendo ovviamente tutte le righe della prima e della seconda colonna
inData_f = np.loadtxt(out_file_f)
x2_f = inData_f[:,0] # Nota Python comincia da "0"!
f2 = inData_f[:,1]
inData_g = np.loadtxt(out_file_f)
x2_g = inData_g[:,0] # Nota Python comincia da "0"!
g2 = inData_g[:,1]

plt.rcParams.update({'font.size': 18})
plt.xlabel("Tempo")
plt.ylabel("Posizione")
plt.title("Funzioni")
plt.plot(x2_f, f2, label="f(x)", color="red")
plt.plot(x2_g, g2, label="g(x)", color="black")
plt.legend(loc="upper left")

filename='figures/funzioni.png'
plt.savefig(filename)
plt.show()
# oppure "upper right", "center", etc
```
